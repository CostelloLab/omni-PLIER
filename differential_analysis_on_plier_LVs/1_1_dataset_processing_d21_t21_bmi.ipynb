```python
#----------Read HTP Transcriptomic dataset for T21 patients in disease co-occurrence conditions -------------------- 
import pandas as pd
master_htp_tpm_w_ds_co_occu = pd.read_csv("./input/master_htp_tpm_w_ds_co_occu.csv", sep=",")
master_htp_tpm_w_ds_co_occu.index=master_htp_tpm_w_ds_co_occu['LabID']

#pro_HTP_WholeBlood_RNAseq_FPKMs_T21.index=pro_HTP_WholeBlood_RNAseq_FPKMs_T21['LabID']

master_htp_tpm_w_ds_co_occu.head(2)
```

```python
meta_data_htp=master_htp_tpm_w_ds_co_occu.iloc[:,range(1,25)].copy()
```

```python
meta_data_htp.head(2)
```

```python
def f(row):
    if row['BMI'] >=30:
        val = 'BMI_GT_30'
    elif row['BMI'] <30:
        val = 'BMI_LT_30'
       
    return val
meta_data_htp = meta_data_htp[meta_data_htp['BMI'].notna()]
meta_data_htp['BMI_dis'] = meta_data_htp.apply(f, axis=1)
display(meta_data_htp.head(2))

print(meta_data_htp.shape)

meta_data_htp.to_csv("./output/HTP_meta_data_D21_T21_BMI.csv", index=False)
```

```python
import pandas as pd

df_htp_plier_model_k0_75_frac1=pd.read_csv("./input/htp_plier_model_k0.75_frac1.csv")



df_htp_plier_model_k0_75_frac1.index=df_htp_plier_model_k0_75_frac1['Unnamed: 0']

df_htp_plier_model_k0_75_frac1=df_htp_plier_model_k0_75_frac1.loc[meta_data_htp.index,]

df_htp_plier_model_k0_75_frac1.head(2)
```

```python
df_htp_plier_model_k0_75_frac1=df_htp_plier_model_k0_75_frac1.drop(['Unnamed: 0'], axis=1)

#df_htp_plier_model_k0_75_frac1=df_htp_plier_model_k0_75_frac1.abs()


#df_htp_plier_model_k0_75_frac1=df_htp_plier_model_k0_75_frac1.apply(pd.to_numeric)



tran_df_htp_plier_model_k0_75_frac1 = df_htp_plier_model_k0_75_frac1.transpose()
tran_df_htp_plier_model_k0_75_frac1['LVs_long']=tran_df_htp_plier_model_k0_75_frac1.index

tran_df_htp_plier_model_k0_75_frac1[['LVs','LVs_name']] = tran_df_htp_plier_model_k0_75_frac1['LVs_long'].str.split(',',expand=True)

tran_df_htp_plier_model_k0_75_frac1['LVs'] = 'LV' + tran_df_htp_plier_model_k0_75_frac1['LVs'].astype(str)


tran_df_htp_plier_model_k0_75_frac1=tran_df_htp_plier_model_k0_75_frac1.drop(['LVs_long','LVs_name'], axis=1)


display(df_htp_plier_model_k0_75_frac1.head(10))

tran_df_htp_plier_model_k0_75_frac1.to_csv('./output/plier_LVS_D21_T2_BMI_htp_for_deseq2.csv', index=False)
```
