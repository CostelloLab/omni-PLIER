```python

import matplotlib.pyplot as plt
import numpy as np
from scipy import stats
import seaborn as sns

import time
import os

import pandas as pd

start_seconds = time.time()

project_root_path=os.path.abspath(os.path.join(os.path.dirname('__file__')))
out_ml_class_dif_co_occu= os.path.join(project_root_path,'output','res_logit_elastic_net_ml_classi_t21_d21_dis_co_occu.csv')

ml_results=pd.read_csv(out_ml_class_dif_co_occu, sep="\t")


dis_co_occu_in_T21= ml_results['dis_co_occu'].unique()
res_plot=os.path.join(project_root_path,'output','htp_tpm_ml_model_per_plier_t21_d21_Karyotype_w_bc.jpeg')


for dco in dis_co_occu_in_T21:
    #print("#############################################")

    dco_fn = dco.replace("/", "")
    dco_fn = dco_fn.replace(" ", "_")
    print(dco_fn)

    best_cv_test_balanced_accuracy_ori_leb=ml_results['test_balanced_accuracy'][(ml_results['lebel_type'] == 'ori_leb') & (ml_results['dis_co_occu'] == dco)]
    best_cv_test_balanced_accuracy_shuff_leb=ml_results['test_balanced_accuracy'][(ml_results['lebel_type'] == 'shuff_leb') & (ml_results['dis_co_occu'] == dco)]
    # Conduct the Kruskal-Wallis Test 
    res_Kruskal_Wallis_Test = stats.kruskal(best_cv_test_balanced_accuracy_ori_leb, best_cv_test_balanced_accuracy_shuff_leb)
    k_p_value=round(res_Kruskal_Wallis_Test[1],3)
    plt.rcParams["figure.figsize"] = [6, 4]
    plt.rcParams["figure.autolayout"] = True
    plt.rcParams["font.family"] = "Arial"
    # plotting first histogram
    ax=sns.histplot(best_cv_test_balanced_accuracy_ori_leb, kde=True)
    # plotting second histogram
    ax=sns.histplot(best_cv_test_balanced_accuracy_shuff_leb, kde=True)
    plt.ylabel("Total Frequency", fontsize=14, fontweight='bold')
    plt.xlabel("Balanced Accuracy on 20% Test Data", fontsize=14, fontweight='bold')
    
    if k_p_value<0.001:
        text_kwalli_test="Kruskal-Wallis Test p<0.001"
    else:
        text_kwalli_test="Kruskal-Wallis Test p>0.001"
        
    ax.annotate(f"D21/T21 - Karyotype on PLIER-LVs", xy=(1.0, 1.0),xycoords='axes fraction', fontsize=14, horizontalalignment='right', verticalalignment='top', fontweight='bold')
    ax.annotate(f"{text_kwalli_test}", xy=(1.0, 0.90),xycoords='axes fraction', fontsize=14, horizontalalignment='right', verticalalignment='top', fontweight='bold')
    
    # Remove the top and right spines
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    plt.xticks(fontsize=12,fontweight='bold')
    plt.yticks(fontsize=12,fontweight='bold')
    plt.savefig(res_plot, bbox_inches = 'tight', dpi=300)
    plt.show()
    
```
