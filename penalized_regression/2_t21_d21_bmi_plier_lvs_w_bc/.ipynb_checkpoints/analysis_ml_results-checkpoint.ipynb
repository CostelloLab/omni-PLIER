```python
# importing libraries
import matplotlib.pyplot as plt
import numpy as np
from scipy import stats
import seaborn as sns

import time
import os

import pandas as pd

start_seconds = time.time()

project_root_path=os.path.abspath(os.path.join(os.path.dirname('__file__'),'..','..'))


#wd_sn_costello_lab/project/htp_da/ml_dl_pipelines/logistic_elastic_net/phase4/t21_w_chr21_w_pro_codi_dis_co_occu/input/master_htp_tpm_w_ds_co_occu.csv
#project_root_path=os.path.join(project_root_path,'t21_w_chr21_w_pro_codi_dis_co_occu','bc_source_sex_age')

#print(project_root_path)

#out_ml_class_dif_co_occu= os.path.join(project_root_path,'output','res_logit_elastic_net_ml_classi_t21_w_chr21_w_pro_codi_dis_co_occu.csv')

out_ml_class_dif_co_occu="./output/res_logit_elastic_net_ml_classi_t21_d21_dis_co_occu.csv"
ml_results=pd.read_csv(out_ml_class_dif_co_occu, sep="\t")
#display(ml_results)

dis_co_occu_in_T21= ml_results['dis_co_occu'].unique()


for dco in dis_co_occu_in_T21:
    #print("#############################################")

    dco_fn = dco.replace("/", "")
    dco_fn = dco_fn.replace(" ", "_")
    print(dco_fn)

    best_cv_test_balanced_accuracy_ori_leb=ml_results['test_balanced_accuracy'][(ml_results['lebel_type'] == 'ori_leb') & (ml_results['dis_co_occu'] == dco)]
    best_cv_test_balanced_accuracy_shuff_leb=ml_results['test_balanced_accuracy'][(ml_results['lebel_type'] == 'shuff_leb') & (ml_results['dis_co_occu'] == dco)]


    # Conduct the Kruskal-Wallis Test 
    res_Kruskal_Wallis_Test = stats.kruskal(best_cv_test_balanced_accuracy_ori_leb, best_cv_test_balanced_accuracy_shuff_leb)

    # Print the result
    k_p_value=round(res_Kruskal_Wallis_Test[1],3)
    
    
    plt.rcParams["figure.figsize"] = [6, 4]
    plt.rcParams["figure.autolayout"] = True


    # plotting first histogram
    ax=sns.histplot(best_cv_test_balanced_accuracy_ori_leb, kde=True)

    # plotting second histogram
    ax=sns.histplot(best_cv_test_balanced_accuracy_shuff_leb, kde=True)

    plt.xlabel("Test Balanced Accuracy")

    #plt.text(0,10,dco, fontsize = 12)
    #ax.text(1, 1, "Custom text", ha="right", va="top")
    
    
    ax.annotate(dco, xy=(0.99, 0.99),xycoords='axes fraction', fontsize=10, horizontalalignment='right', verticalalignment='top')
    ax.annotate(f"kruskal p={k_p_value}", xy=(0.95, 0.95),xycoords='axes fraction', fontsize=10, horizontalalignment='right', verticalalignment='top')
    
    plt.show()

    #plt.xlim(0.2, 0.8)

    #plt.text(0, 0, 'matplotlib', horizontalalignment='center',verticalalignment='center')

    # Showing the plot using plt.show()
    #plt.show()

```
