```python
import gseapy as gp
from gseapy import Msigdb
msig = Msigdb()
# mouse hallmark gene sets
#gmt = msig.get_gmt(category='c2.all', dbver="2023.2.Hs")
#gmt='/sutanu/media/wd_sn_costello_lab/project/htp_da/reverse_eng_PLIER_to_MSigDB/input/data/unzip/msigdb_v4.0_files_to_download_locally/msigdb_v4.0_GMTs/c2.all.v4.0.symbols.gmt'
gmt='./input/msigdb_v_4_C2_c6_c7_bloodCellMarkersIRISDMAP_svmMarkers.gmt'
gsea_plot="./output/htp_tpm_gsea_c2_c6_c7_bloodCellMarkersIRISDMAP_svmMarkers_all_pro_cod_t21_obesity.jpeg"
# run gsea
# enrichr libraries are supported by gsea module. Just provide the name
gs_res = gp.gsea(data=tran_htp_tpm_pro_cod_t21_obesity_remp1, # or data='./P53_resampling_data.txt'
                 gene_sets=gmt, # or enrichr library names
                 cls= class_vector, # cls=class_vector
                 # set permutation_type to phenotype if samples >=15
                 threads=25,
                 min_size=5,
                 max_size=500,
                 permutation_num=100000, # reduce number to speed up testing
                 outdir=None, # don't write to disk
                 method='signal_to_noise',
                 seed=6,
                 verbose=True, # see what's going on behind the scenes
                )
gs_res.pheno_pos = "Obesity +Ve"
gs_res.pheno_neg = "Obesity -Ve"


terms = gs_res.res2d.Term
axs = gs_res.plot(terms=terms[1])


axs.savefig(gsea_plot, bbox_inches = 'tight')


gs_res.res2d.to_csv("./output/htp_tpm_gsea_c2_c6_c7_bloodCellMarkersIRISDMAP_svmMarkers_all_pro_cod_t21_obesity.csv")


gsea_ranking=gs_res.ranking
genes_w_ranking=pd.DataFrame(gsea_ranking)
genes_w_ranking['gene_name']=genes_w_ranking.index
genes_w_ranking=genes_w_ranking.rename(columns={0:'Ranks'})


genes_w_ranking.to_csv("./output/rank_htp_tpm_gsea_c2_c6_c7_bloodCellMarkersIRISDMAP_svmMarkers_all_pro_cod_t21_obesity.csv")

display(gs_res.res2d.head(10))
display(gs_res.res2d.tail(10))
    
```
